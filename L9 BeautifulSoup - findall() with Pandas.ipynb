```python
# importing library
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r = requests.get(url)
r.status_code
```

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
# lets get all names of products
p_name = soup.find_all("a" , class_ = "title")
```

```python
product_name = [ ]
for i in p_name:
    name = i.text
    product_name.append(name)
product_name
```

```python
# now  for price
p_price = soup.find_all("h4",class_="pull-right price")
product_price = [ ]
for i in p_price:
    price = i.text
    product_price.append(price)

product_price
```

```python
# lets take description 
p_dec = soup.find_all("p" ,class_ = "description")
product_dec = [ ]
for i in p_dec:
    dec = i.text
    product_dec.append(dec)
    
product_dec
```

```python
# get reviews
p_rew = soup.find_all("p",class_ = "pull-right")
product_rev = [ ]
for i in p_rew:
    rew = i.text
    product_rev.append(rew)
product_rev
```

```python
# # try for  rating
p_rating = soup.find_all("div",class_ = "ratings") 
# only_rate = p_rating.find_all("p")



# rating = []
# for i in p_rating:
#     rate = i.text
#     rating.append(rate)
# rating

p_rating
```

```python
# lets malke dataframe from it
data = pd.DataFrame({"product_name":product_name,"description":product_dec,"price":product_price,"reviews":product_rev})
```

```python
data.to_csv("laptop_info.csv")
```
