```python
import requests
from bs4 import BeautifulSoup

for i in range(2, 21):
    url = "https://www.flipkart.com/search?q=smartphone+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + str(i)
    re = requests.get(url)
    soup = BeautifulSoup(re.text, "lxml")
    np = soup.find("a", class_="_1LKTO3").get("href")
    print("https://www.flipkart.com" + np)
```
