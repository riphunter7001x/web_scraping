```python
# importing libraryequipped 
import requests
from bs4 import BeautifulSoup
```

```python
#get url 
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r = requests.get(url)
r.status_code

soup = BeautifulSoup(r.text,"lxml")
```

```python
price = soup.find_all("h4", class_="pull-right price")

for i in price:
    print(i.text)
```

```python
# let get discription
desc = soup.find_all("p", class_="description")

for i in desc:
    print(i.text)
```

```python
# lets print together

for i in price:
    for j in desc:
        print(i.text)
        print(j.text)
```

```python
# another  way
for i, j in zip(price, desc):
    print(i.text)
    print(j.text)
```
