```python
# import library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# getting points tablle Most Runs
url = "https://www.mykhel.com/cricket/ipl-stats-s4/"
r = requests.get(url)
r.status_code
```

```python
# loading soup
soup = BeautifulSoup(r.text,"lxml")
soup
```

```python
# getting table titles
table = soup.find("table",class_ = "cricket-performanceTable t20")
table
```

```python
# getting headers
header = table.find_all("tr",class_ = "mainbgcolor")
```

```python
# lets add it in 
title = []
for i in header:
    title.append(i.text)
title

string = title[0]
titles = string.split()
titles
```

```python
# add it in data frame
df = pd.DataFrame(columns = titles)
```

```python
df
```

```python
# now lets get player data
player = table.find_all("tr")
```

```python
# get data
for i in player[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    new_list = [item for item in row if item]
#     print(new_list)
    l = len(df)
    df.loc[l] = new_list
```
