```python
# importing library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://ticker.finology.in/market/52-week-high"
r = requests.get(url)
# checking data is  connected or not
r.status_code
```

```python
# make soup 
soup = BeautifulSoup(r.text,"lxml")
```

```python
table = soup.find("table")
```

```python
header = table.find_all("th")
```

```python
headers= []
for i in header:
    headers.append( i.text)
```

```python
headers
```

```python
# lest make data farme

df = pd.DataFrame(columns=headers)
```

```python
df
```

```python
#  now add data in tble

rows = table.find_all("tr")
rows
```

```python
# we start from index 1 becoz at zero we alredy have our  titles
for i in rows[1:]:
    data = i.find_all("td")
#     print(row)
    row = [tr.text for tr in data]
#     print(row)
#     add in our df
    l = len(df)
    df.loc[l] = row
    
    
```

```python
# print df
df
```

```python
df.to_csv("stock_market_data.csv")
```
